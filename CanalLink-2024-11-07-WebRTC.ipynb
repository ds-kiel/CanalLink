```python
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

FIG_SIZE=(12, 4)
```

```python
df_starlink = pd.read_csv('data/2024-11-07-webrtc/starlink_interpolated.csv')
df_starlink['timestamp'] = pd.to_datetime(df_starlink['timestamp'])
df_starlink.set_index('timestamp', inplace=True, drop=False)
df_starlink.info()
```

```python
#df_starlink['rtt_ms'].plot(figsize=(40, 10), ylim=(0, 1000))
#df_vodafone['rtt_ms'].plot(figsize=(40, 10), ylim=(0, 1000))

fig, ax = plt.subplots(figsize=FIG_SIZE)

df_starlink['rtt_ms'].rename('Starlink-RTT').plot(ax=ax, ylim=(0, 1000))
df_vodafone['rtt_ms'].rename('Vodafone-RTT').plot(ax=ax, ylim=(0, 1000))

ax.set_title('RTT over time')
ax.set_ylabel('RTT [ms]')
ax.grid(True, which='both', linestyle='--', alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
#df_starlink['jitter_ms'].plot(figsize=(40, 10), ylim=(0, 50))
#df_vodafone['jitter_ms'].plot(figsize=(40, 10), ylim=(0, 50))

fig, ax = plt.subplots(figsize=FIG_SIZE)

df_starlink['jitter_ms'].rename('Starlink-Jitter').plot(ax=ax, ylim=(0, 100))
df_vodafone['jitter_ms'].rename('Vodafone-Jitter').plot(ax=ax, ylim=(0, 100))


ax.axhline(df_starlink['jitter_ms'].mean(), linestyle='--', alpha=0.7, color='tab:blue', label='Starlink Mean')
ax.axhline(df_vodafone['jitter_ms'].mean(), linestyle=':', alpha=0.7, color='tab:red', label='Vodafone Mean')

ax.set_title('Jitter over time')
ax.set_ylabel('Jitter [ms]')
ax.grid(True, which='both', linestyle='--', alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16,8), dpi=200)

# Primary axis: tx rates
df_starlink['tx_rate_mbits'].rename('Starlink-Tx-Rate').plot(ax=ax, ylim=(0, 7), color='tab:blue')
df_vodafone['tx_rate_mbits'].rename('Vodafone-Tx-Rate').plot(ax=ax, ylim=(0, 7), color='tab:red')

# # Means on primary axis
# ax.axhline(df_starlink['tx_rate_mbits'].mean(),
#            linestyle='--', alpha=0.7, label='Starlink Mean')
# ax.axhline(df_vodafone['tx_rate_mbits'].mean(),
#            linestyle=':', alpha=0.7, label='Vodafone Mean')

ax.set_title('Tx rate over time')
ax.set_ylabel('Rate [Mbps]')
ax.grid(True, which='both', linestyle='--', alpha=0.3)

# Secondary axis: cell IDs
ax2 = ax.twinx()

df_starlink['rtt_ms'].rename('Starlink-RTT').plot(ax=ax2, ylim=(0, 1000), alpha=0.7, color='tab:green')
df_vodafone['rtt_ms'].rename('Vodafone-RTT').plot(ax=ax2, ylim=(0, 1000), alpha=0.7, color='tab:orange')
ax2.set_ylabel('RTT [ms]')

# Combine legends from both axes
handles1, labels1 = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax.legend(handles1 + handles2, labels1 + labels2, loc='upper left')

plt.tight_layout()
plt.show()
```
